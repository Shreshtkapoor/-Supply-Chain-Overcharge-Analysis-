```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
import warnings
warnings.filterwarnings("ignore")
```

```python
order_report = pd.read_excel('Company X - Order Report.xlsx')
courier_invoice = pd.read_excel('Courier Company - Invoice.xlsx')
sku_master = pd.read_excel('Company X - SKU Master.xlsx')
courier_rates = pd.read_excel('Courier Company - Rates.xlsx')
pincodes = pd.read_excel('Company X - Pincode Zones.xlsx')
```

```python
order_report.rename(columns={'ExternOrderNo' : 'Order ID'}, inplace=True)
```

```python
order_merged = order_report.merge(courier_invoice, on="Order ID")
```

```python
order_merged
```

```python
order_sku = order_merged.merge(sku_master,on = 'SKU')
```

```python
order_sku[order_sku.duplicated()]
```

```python
order_sku = order_sku.drop_duplicates()
```

```python
order_sku
```

```python
order_sku['Weight'] = order_sku['Weight (g)']/ 1000
```

```python
order_sku.rename(columns = {'Weight' : 'Cp_Weight'}, inplace=True)
```

```python
order_sku.drop(columns = 'Weight (g)',inplace=True)
```

```python
order_sku['Cp_Weight'] = order_sku['Cp_Weight'].round(2)
```

```python
courier_rates
```

```python
# Convert Zone to lowercase for consistency
order_sku['Zone'] = order_sku['Zone'].str.lower()

# Assign fixed and additional rates based on the Zone
order_sku['fwd_fixed'] = order_sku['Zone'].apply(lambda zone: courier_rates.at[0, f'fwd_{zone}_fixed'])
order_sku['fwd_additional'] = order_sku['Zone'].apply(lambda zone: courier_rates.at[0, f'fwd_{zone}_additional'])
order_sku['rto_fixed'] = order_sku['Zone'].apply(lambda zone: courier_rates.at[0, f'rto_{zone}_fixed'])
order_sku['rto_additional'] = order_sku['Zone'].apply(lambda zone: courier_rates.at[0, f'rto_{zone}_additional'])

# Calculate Forward Additional Charges based on weight slabs
order_sku['Total_Forward_Additional'] = order_sku.apply(
    lambda row: math.ceil((row['Cp_Weight'] - 0.5) / 0.5) * row['fwd_additional'] if row['Cp_Weight'] > 0.5 else 0, axis=1
)

# Calculate Total Forward Charges
order_sku['Total_fwd_charges'] = order_sku['fwd_fixed'] + order_sku['Total_Forward_Additional']

# Calculate RTO Additional Charges based on weight slabs (if applicable)
order_sku['Total_RTO_Additional'] = order_sku.apply(
    lambda row: math.ceil((row['Cp_Weight'] - 0.5) / 0.5) * row['rto_additional'] if row['Type of Shipment'] == 'Forward and RTO charges' and row['Cp_Weight'] > 0.5 else 0, axis=1
)

# Calculate Total RTO Charges (if applicable)
order_sku['Total_RTO_Charges'] = order_sku.apply(
    lambda row: row['rto_fixed'] + row['Total_RTO_Additional'] if row['Type of Shipment'] == 'Forward and RTO charges' else 0, axis=1
)

# Calculate the Final Actual Rates
order_sku['Actual_Rates'] = (order_sku['Total_fwd_charges'] + order_sku['Total_RTO_Charges'])*order_sku['Order Qty']
```

```python
Final_Report['Courier Slab(X)'] = Final_Report['Cp_Weight'].apply(lambda x: math.ceil(x / 0.5))

column_order = Final_Report.columns.tolist()  # Get current column order
column_order.insert(column_order.index('Cp_Weight') + 1, column_order.pop(column_order.index('Courier Slab(X)')))

Final_Report = Final_Report[column_order]
```

```python
Final_Report['Courier Slab(C)'] = Final_Report['Charged Weight'].apply(lambda x: math.ceil(x/0.5))

column_order1 = Final_Report.columns.tolist()
column_order1.insert(column_order1.index('Charged Weight') + 1, column_order1.pop(column_order1.index('Courier Slab(C)')))
Final_Report = Final_Report[column_order1]
```
